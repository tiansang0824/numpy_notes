# 案例：图片翻转


```python
import numpy as np
```

```python
# 一个简单的翻转测试

n = np.random.randint(0, 100, size=(5,6))
n
```

```python
# 行翻转
n[::-1]
```

通过上面两个代码块可以看出，直接使用切片操作的[::-1]，将会对矩阵的行进行翻转。

```python
# 列翻转
n[:,::-1 ]
```

列翻转，相当于是对第二个维度进行翻转。

如果有第三个维度，也可以用这种方法翻转第三维度的数据。


```python
from matplotlib import pyplot as plt

# 对图片进行翻转
ellen_joe = plt.imread("./resources/ellen_joe.jpg")  # 读取图片
ellen_joe.shape
```

```python
# 显示原图，等会儿作对比
plt.imshow(ellen_joe)
```

```python
# 对行翻转，就是上下翻转
plt.imshow(ellen_joe[::-1])
```

```python
# 对列翻转就是图形的左右翻转
plt.imshow(ellen_joe[:,::-1])
```

```python
# 对第三维度进行翻转：rgb->bgr
plt.imshow(ellen_joe[:,:,::-1])
```

```python
# 通过增大步长，可以对图片进行模糊处理
plt.imshow(ellen_joe[::10,::10])
```
